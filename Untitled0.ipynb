```python
import torch
from transformers import PegasusForConditionalGeneration, PegasusTokenizer

device = "cuda" if torch.cuda.is_available() else "cpu"

model_name = "google/pegasus-xsum"
tokenizer = PegasusTokenizer.from_pretrained(model_name)
model = PegasusForConditionalGeneration.from_pretrained(model_name).to(device)

print(f"Model loaded on: {device}")
```

```python
def summarize(text, num_beams=4, length_penalty=1.0, max_length=64):
    # 1. Convert article text into token IDs
    inputs = tokenizer(text, truncation=True, max_length=512,
                       return_tensors="pt").to(device)

    # 2. Generate summary token IDs
    summary_ids = model.generate(
        **inputs,
        num_beams=num_beams,
        length_penalty=length_penalty,
        max_length=max_length,
        no_repeat_ngram_size=3,
        early_stopping=True,
    )

    # 3. Convert token IDs back into readable text
    return tokenizer.decode(summary_ids[0], skip_special_tokens=True)


def word_count(text):
    return len(text.split())

print("Helper functions ready ✅")
```

```python
#PART D--
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.text_rank import TextRankSummarizer

def textrank_summarize(text, n_sentences=2):
    parser = PlaintextParser.from_string(text, Tokenizer("english"))
    summarizer = TextRankSummarizer()
    sentences = summarizer(parser.document, n_sentences)
    return " ".join(str(s) for s in sentences)

for name, text in articles.items():
    print(f"===== {name} =====")
    print("PEGASUS (abstractive):")
    print("  ", pegasus_summaries[name])
    print("TEXTRANK (extractive):")
    print("  ", textrank_summarize(text))
    print()
```
